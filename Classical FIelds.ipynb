```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#a

plt.figure(figsize=(16,10))
y=0
x=np.linspace(-2.5,3.3,100)
h=np.exp(-x**2-y**2)+0.5*np.exp(-(x-1.5)**2-2*(y**2))
plt.plot(x,h,'c')
plt.title('Plot of $h$ at $y = 0$')

```

```python
# b

plt.figure(figsize=(12,10))
wyg= np.linspace(-2,2,100)
wxg= np.linspace(-2,3,100)
xg,yg=np.meshgrid(wxg,wyg)
hg=np.exp(-xg**2-yg**2)+0.5*np.exp(-(xg-1.5)**2-2*(yg**2))
contourplot = plt.contour(xg,yg,hg,np.arange(0.1,1.1,0.1),cmap='Paired')
plt.axis('equal')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Contours')
plt.clabel(contourplot)
```

```python
#d

plt.figure(figsize=(16,10))
wyg= np.linspace(-2,2,100)
wxg= np.linspace(-2,3,100)
xg,yg=np.meshgrid(wxg,wyg)
hg=np.exp(-xg**2-yg**2)+0.5*np.exp(-(xg-1.5)**2-2*(yg**2))
ax=plt.axes()
ax.contour(xg,yg,hg,np.arange(0.1,1.1,0.1),cmap='Paired')
ax.set_aspect('equal')

wyg1= np.arange(-2,2,0.18)
wxg1= np.arange(-2,3,0.24)
xg1,yg1=np.meshgrid(wxg1,wyg1)
gradhx=-2*xg1*np.exp(-xg1**2-yg1**2)-(np.exp(-((xg1-1.5)**2)-2*yg1**2))*(xg1-1.5)
gradhy=-2*yg1*np.exp(-xg1**2-yg1**2)-(np.exp(-((xg1-1.5)**2)-2*yg1**2))*2*yg1
ax.quiver(xg1,yg1,gradhx,gradhy,color='m',width=0.0015)
ax.set_aspect('equal')
ax.set_title('Quivers and Contour Plot using analytical grad(h)')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
```

```python
#e

plt.figure(figsize=(16,10))
wyg= np.linspace(-2,2,100)
wxg= np.linspace(-2,3,100)
xg,yg=np.meshgrid(wxg,wyg)
hg=np.exp(-xg**2-yg**2)+0.5*np.exp(-(xg-1.5)**2-2*(yg**2))
ax1=plt.axes()
ax1.contour(xg,yg,hg,np.arange(0.1,1.1,0.1),cmap='Paired')
ax1.set_aspect('equal')

h1 = np.exp(-xg1**2-yg1**2)+0.5*np.exp(-(xg1-1.5)**2-2*(yg1**2))
gradhy1 , gradhx1 = np.gradient(h1,0.5)
ax1.quiver(xg1,yg1,gradhx1,gradhy1,color='k',width=0.0015)
ax1.set_title('Quivers and Contour Plot using Numpy"s gradient function')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
```

```python
#f with loop

fig=plt.figure(figsize=(24,14))         #Figure window

wyg= np.linspace(-2,2,50)               #Ranges Defined for contours
wxg= np.linspace(-2,3,50)
xg,yg=np.meshgrid(wxg,wyg)              #grid defined over ranges using meshgrid
hg=np.exp(-xg**2-yg**2)+0.5*np.exp(-(xg-1.5)**2-2*(yg**2))                             #field calculated over grid values

ax=plt.axes()                           #axes defined
ax.set_xlabel('$x$')                    #axes labels
ax.set_ylabel('$y$')
ax.set_title('Quivers on top of Contours plot')                                          #plot title
ax.set_aspect('equal')                  #equal aspect ratio so plot does not look squishy,elongated

colormaps = ['BrBG','tab20b','cool','Accent','Pastel1','rainbow','twilight','Dark2','Wistia','Paired']   

                                        # List of 10 colormaps to be used since for each contour value (0.1 - 1) because 
                                        # if only a single colormap was used then all the contours would be of same colour
                                        # as the contour plotting function is plotting a single contour in each iteration of 
                                        # for loop (it chooses the least bright colour of a colormap )


i=0                                     #index defined for the first string in the list colormaps

for cval in np.arange(0.1,1.1,0.1):                                 #contour value range defined,chaning each iteration
    c1 = ax.contour(xg,yg,hg,[cval],cmap=colormaps[i])              #plotting contour using initial cval and color map
    c1array = c1.collections[0].get_paths()[0].vertices             #extracting the pair of xgrid and ygrid  coordinate
                                                                    #arrays for the cval value
    
    c1x = c1array[:,0]                                     #extracting the x grid array values only
    c1y = c1array[:,1]                                     #extracting the y grid array values only
    wxg1 = c1x                                             #defining the grids for quiver plots, no need to use meshgrid 
    wyg1 = c1y                                             #again since extracted arrays from contours were already 
    xg1 = wxg1                                             #'meshgrid-ed'(when plotting contours)
    yg1 = wyg1
    gradhx1 = -2*xg1*np.exp(-xg1**2-yg1**2)-(np.exp(-((xg1-1.5)**2)-2*yg1**2))*(xg1-1.5) #calculating grad values over the grid
    gradhy1 = -2*yg1*np.exp(-xg1**2-yg1**2)-(np.exp(-((xg1-1.5)**2)-2*yg1**2))*2*yg1
    
    ax.quiver(xg1,yg1,gradhx1,gradhy1,color='r',width=0.001, angles='xy', scale =50)     #using analytically calculated grad
                                                                                         #values and quiver plotting
    
    
    i= i + 1                        #updating the string index in the
                                    #list colourmaps
```
